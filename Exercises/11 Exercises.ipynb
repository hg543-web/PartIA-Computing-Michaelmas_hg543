## Exercise 11.2

For the recursive factorial algorithm in Activity 04, determine the algorithmic complexity by inspecting your implementation of the algorithm. Test this against numerical experiments.

### Solution

Recall the factorial algorithm from Activity 04.4:

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return factorial(n - 1)*n
```

The function calls itself (recursively) $n$ times, hence it has complexity $O(n)$. We test this below and plot
the times.

```python
...
n = range(20)
t = [] # empty list for the times
for i in n:
    time = %timeit -n1 -o -q factorial(i) # calculates time for each function
    t.append(time.best)
plt.plot(n, t, marker='o', label = r'$O(n!)$')
plt.plot(n, [5.0e-8*i for i in n], '--', label= r'n') # Adjusted scaling so relationship is more noticeable
plt.legend(loc=0) # places plot in top left

plt.xlabel('n')
plt.ylabel('t/s')

plt.show() # supports relationship
```

## Exercise 11.3

Determine experimentally the complexity of computing the determinant of a matrix. You can generate an $n \times n$ matrix using:

```python
n = 100
A = np.random.rand(n, n)
```

and the determinant can be computed by:

```python
det = np.linalg.slogdet(A)
```

Be sure that you test for sufficiently large $n$ to get into the 'large' $n$ regime.

### Solution

Time computation of determinant:

```python
# Create array of problem sizes we want to test (powers of 2)
N = 2**np.arange(2, 12)
 
...

times = []

for n in N:
    A = np.random.rand(n, n)
    t= %timeit -q -n4 -r1 -o np.linalg.slogdet(A)
    times.append(t.best) # adds computational time (complexity) to a list
```

Plot result:

```python
...
plt.plot(N, times, marker='o', label='Determinent')
plt.xlabel('n')
plt.ylabel('t/s')

plt.legend(loc=0)

plt.show()

plt.plot(N**3, times, marker='o', label='Determinent')
# n**3 complexity since it is effectively a matrix-matrix multiplication
plt.xlabel('n**3')
plt.ylabel('t/s')

plt.legend(loc=0)

plt.show()
```
